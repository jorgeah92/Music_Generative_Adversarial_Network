```python
from mido import MidiFile, MidiTrack, Message
import pandas as pd
import math
```

```python
mid = MidiFile('All_Maestro/ORIG-MIDI_02_7_7_13_Group__MID--AUDIO_17_R1_2013_wav--4.midi') 
```

```python
message_list = []
for i in mid.tracks[1][2:-1]: 
    message_list.append(i) 
message_list
```

```python
message_strings = []
for x in message_list:
    message_strings.append(str(x))
message_strings
```

```python
message_strings_split = []
for message in message_strings:  
    split_str = message.split(" ")
    message_strings_split.append(split_str)
message_strings_split
```

```python
message_type = []
for item in message_strings_split:
    message_type.append(item[0])
df1 = pd.DataFrame(message_type)
df1
```

```python
attributes = []
for item in message_strings_split:
    attributes.append(item[1:])
attributes_dict = [{}]    
for item in attributes:
    for i in item:
        key, val = i.split("=")
        if key in attributes_dict[-1]:
            attributes_dict.append({})
        attributes_dict[-1][key] = val
df2 = pd.DataFrame.from_dict(attributes_dict)
df2
```

```python
df_complete = pd.concat([df1, df2], axis=1)
df_complete.time = df_complete.time.astype(float)
# create a time elapsed attribute equal to the cumulative sum of time.
df_complete['time_elapsed'] = df_complete.time.cumsum()
df_complete
```

```python
# df_complete[df_complete[0] == 'control_change']['control'].drop_duplicates()
# control change messages are for the pedal...let's simplify by not having those:
df_notes = df_complete[df_complete[0] == 'note_on'].drop(columns={0,'channel','control','value'}).reset_index(drop=True)
df_notes.note = df_notes.note.astype(int)
df_notes.velocity = df_notes.velocity.astype(int)
df_notes
```

```python
# find the duration each note was played based on the stop note signals (note on with velocity == 0)
duration = [0] * len(df_notes)
for i in range(len(df_notes)):
    if df_notes['velocity'][i] != 0:
        j = i + 1
        while df_notes['note'][j] != df_notes['note'][i]:
            j += 1
        duration[i] = df_notes['time_elapsed'][j] - df_notes['time_elapsed'][i]
df_notes['duration'] = duration
df_notes_on = df_notes[df_notes['velocity'] != 0].drop(columns={'time'}).reset_index(drop=True)
df_notes_on
```

```python
# simplify again to start without dynamics:
df_notes_only = df_notes_on.drop(columns='velocity')
df_notes_only
```

```python
# say we would work with a 200 note subset of the song, which is over 10 seconds of actual music:
song_len = 200
df_first_200 = df_notes_only.loc[0:song_len-1].copy()
df_first_200
```

```python
# here, this translates to 13-14 seconds of music (a good amount)
# now, let's normalize the time elapsed and make duration a fraction of time elapsed:
df_first_200['time_elapsed'] -= df_first_200['time_elapsed'][0]
df_first_200['duration'] /= df_first_200['time_elapsed'][song_len-1]
df_first_200['time_elapsed'] /= df_first_200['time_elapsed'][song_len-1]
```

```python
# finally, let's recreate the "time since last event" nature of a midi file for time_elapsed:
time_since_last = [0] * song_len
for i in range(1, song_len):
    time_since_last[i] = df_first_200['time_elapsed'][i] - df_first_200['time_elapsed'][i-1]
df_first_200['time_since_last'] = time_since_last
df_first_200
```

```python
df_first_200['note'] -= 20
df_first_200['note'] /= 88
df_first_200 = df_first_200.drop(columns='time_elapsed')
df_first_200
```

```python
# Now we need to go backwards, and create a midi file from the first 200 notes:
# Can start by reverse scaling the note:
df_reversed = df_first_200.copy()
df_reversed['note'] = round(df_reversed['note'] * 88 + 20)  # might want to have something more special than round()
df_reversed.note = df_reversed.note.astype(int)
df_reversed
```

```python
df_reversed['velocity'] = 60  # create a uniform middling velocity
df_reversed
```

```python
# recreate the absolute time index and drop time_since_last (we'll recreate it with the stop signals)
df_reversed['time_index'] = df_reversed.time_since_last.cumsum()
df_reversed = df_reversed.drop(columns = 'time_since_last')
df_reversed
```

```python
# create a stop signal for each note at the appropriate time_index:
for i in range(len(df_reversed)):
    stop_note = pd.DataFrame([[df_reversed.note[i], 0, 0, df_reversed.duration[i] + df_reversed.time_index[i]]],
                             columns=['note', 'duration', 'velocity', 'time_index'])
    df_reversed = df_reversed.append(stop_note, ignore_index=True)
df_reversed = df_reversed.sort_values('time_index').reset_index(drop=True)
df_reversed
```

```python
# recreate time_since last with the stop note signals, and now we don't need duration or time_index:
df_reversed['time'] = [0] + [df_reversed.time_index[i+1] - df_reversed.time_index[i] 
                             for i in range(len(df_reversed)-1)]
df_reversed = df_reversed.drop(columns = {'time_index','duration'})
df_reversed
```

```python
# finally, we need to scale the time since last note appropriately:
# will probably try different values here within a function, since it will be how fast the piece is played
df_reversed['time'] = round(df_reversed['time'] * 20000)
df_reversed.time = df_reversed.time.astype(int)
df_reversed
```

```python
# turn the dataframe back into a midi file, save (along with the initial) for a comparison
mid_remade = MidiFile()
track = MidiTrack()
mid_remade.tracks.append(track)
track.append(Message('program_change', program=0, time=0))
for i in range(len(df_reversed)):
    track.append(Message('note_on', note=df_reversed.note[i], velocity=df_reversed.velocity[i], time=df_reversed.time[i]))

mid.save('mid_initial_test.mid')
mid_remade.save('mid_remade_test.mid')
```
